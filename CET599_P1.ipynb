```python
#parameter 
cell_len = 0.1

#free flow speed(mile/min)
u_f = 0.5
#shockwave speed(mile/min)
w = 0.25
#diverge rate
beta_12 = 0.6
beta_13 = 0.4
#demand(veh/min)
inflow = 10
T_len = 10
```

```python
#set time step
tstep = round(cell_len/u_f,2)
tstep
#cell_len = tstep*u_f
```

```python
#link: [length, capacity, density, Ni for each cell of each link, qi for each cell of each link]
para = {1:[0.5,10,160], 2:[0.5,10,160], 3:[0.5,10,160],4:[1,4,124],5:[1.5,3,118],6:[0.5,10,160]}

#number of cell for each link
no_cell = list()
l = len(para)
for i in range(l):
    no = int(para[i+1][0]/cell_len)
    no_cell.append(no)


    
#Ni, qi for each link
for i in range(1,len(para)+1):
    N = para[i][2]*para[i][0]/(no_cell[i-1])
    q = para[i][1]/(no_cell[i-1])
    para[i].append(N)
    para[i].append(q)
no_cell
col = sum(no_cell)
```

```python
no_step = list()
a = 0
for i in range(len(no_cell)):
    a = a + no_cell[i]
    no_step.append(a)    
```

```python
#set original row
Table1=[[0]]
l = int(20/tstep + 1)
for i in range(1,l):
    a = list()
    t = round(i*tstep,2)
    a.append(t)
    Table1.append(a)
#initialization: input values for n1
Tl = len(Table1)
n1 = Table1
n1[0].append(0)
for i in range(1,Tl):
    if i <= T_len/tstep:
        n = 2
    else:
        n = 0
    n1[i].append(n)
```

```python
def get_alpha(q,n):
    if n <= q:
        return 1
    else:
        return 0.5
```

```python
#for link 1
#for Table of ni 
#initialiaze the first row
for i in range(col-1):
    n1[0].append(0)
```

```python
import statistics

l=len(n1)

y1
for j in range(l-1):
    #calculation for # of veh 
    for i in range(2, col):
        nit = n1[j][i]+y1[j][i]-y1[j][i+1]
        n1[j+1].append(nit)
    n1[j+1].append(0)


    
    for b in range(2, col+1):
        #link 1, 6-1
        if b < no_step[0]:
            q = para[1][4]
            n = n1[j][b]
            alpha = get_alpha(q,n)
            y = min(n1[j+1][b-1], q, alpha*(para[1][3]-n1[j][b]))
        #for node1, diverge
        elif b == no_step[0]:
            sbk = min(para[1][4],n1[j+1][b])
            q1 = para[2][4]

            alpha1 = get_alpha(q1,n1[j][b])
            rek = min(para[2][4],alpha1*(para[2][3]-n1[j][b+1]))
            
             
            alpha2 = get_alpha(para[3][4],n1[j][no_step[1]+1])
            rck = min(para[3][4],alpha2*(para[3][3]-n1[j][no_step[1]+1]))
            y = min(sbk, rek/beta_12, rck/beta_12)
        #for node1 to link 2, 1-2
        elif b == no_step[0]+1:
            y = y1[j+1][no_step[0]]*beta_12
        #for link 1-2
        elif no_step[0]+1 < b <= no_step[1]:
            q = para[2][4]
            n = n1[j][b]
            alpha = get_alpha(q,n)
            y = min(n1[j+1][b-1], q, alpha*(para[2][3]-n1[j][b]))
        #for node1 to link 3, 1-3
        elif b == no_step[1]+1:  
            y = y1[j+1][no_step[0]]*beta_13
        #for link 1-3
        elif no_step[1]+1 <b <= no_step[2]:
            q = para[3][4]
            n = n1[j][b]
            alpha = get_alpha(q,n)
            y = min(n1[j+1][b-1], q, alpha*(para[3][3]-n1[j][b]))
        #for link 2-4
        elif no_step[2]< b<= no_step[3]:
            q = para[4][4]
            n = n1[j][b]
            alpha = get_alpha(q,n)
            y = min(n1[j+1][b-1], q, alpha*(para[4][3]-n1[j][b])) 
        #for link 3-4
        elif no_step[3]<b<= no_step[4]:
            q = para[5][4]
            n = n1[j][b]
            alpha = get_alpha(q,n)
            y = min(n1[j+1][b-1], q, alpha*(para[5][3]-n1[j][b]))
        #for node 4, emerge, 2-4,3-4 to 5
        elif b == no_step[4]+1:
            sbk = min(para[4][4],n1[j+1][b])
            q1 = para[6][4]
             
            alpha1 = get_alpha(q1,n1[j][b])
            rek = min(para[6][4],alpha1*(para[6][3]-n1[j][b+1]))
            sck = min(para[5][4],n1[j+1][no_step[4]])
            y3 = statistics.median([sbk,rek-sck,beta_12*rek])

            sck = min(para[5][4],n1[j+1][b])
            q1 = para[6][4]
            alpha1 = get_alpha(q1,n1[j][b])
            rek = min(para[6][4],alpha1*(para[6][3]-n1[j][b+1]))
            sbhk = min(para[4][4],n1[j+1][no_step[3]])
            y2 = statistics.median([sck,rek-sbk,beta_13*rek])

            y = y3+y2
        #for link 4-5
        else:
            q = para[6][4]
            n = n1[j][b]
            alpha = get_alpha(q,n)
            y = min(n1[j+1][b-1], q, alpha*(para[6][3]-n1[j][b]))  
        y1[j+1].append(y)    

     
```
